```python
from pyecharts import options as opts
from pyecharts.charts import Graph

# 使用之前定义的 data
nodes = data["nodes"]
links = data["links"]
categories = data["categories"]

# 归一化节点大小
node_values = [node["value"] for node in nodes]
min_node_value = min(node_values)
max_node_value = max(node_values)

for node in nodes:
    # 将 value 归一化到 [10, 30] 范围
    normalized_size = (node["value"] - min_node_value) / (max_node_value - min_node_value) * 20 + 10
    node["symbolSize"] = normalized_size  # 动态设置节点大小

# 归一化线条粗细
link_values = [link["value"] for link in links]
min_link_value = min(link_values)
max_link_value = max(link_values)

for link in links:
    # 将 value 归一化到 [1, 6] 范围
    normalized_width = (link["value"] - min_link_value) / (max_link_value - min_link_value) * 5 + 1
    link["lineStyle"] = {"width": normalized_width}  # 动态设置线条粗细

# 创建图
c = (
    Graph(init_opts=opts.InitOpts(width="1920px", height="1080px", bg_color="#000"))  # 设置背景颜色为黑色
    .add(
        "",
        nodes=nodes,
        links=links,
        categories=categories,
        layout="circular",
        is_rotate_label=True,
        linestyle_opts=opts.LineStyleOpts(color="source", curve=0.3),
        label_opts=opts.LabelOpts(position="right", color="#fff"),  # 设置节点标签字体颜色为白色
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Graph-University and Country Connections", title_textstyle_opts=opts.TextStyleOpts(color="#fff")),  # 设置标题字体颜色为白色
        legend_opts=opts.LegendOpts(
            orient="vertical",
            pos_left="2%",
            pos_top="20%",
            textstyle_opts=opts.TextStyleOpts(color="#fff"),  # 设置图例字体颜色为白色
        ),
    )
)

# 在Jupyter Notebook中显示图表
c.render_notebook()
```
